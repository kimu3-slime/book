## 1.1 数式の基本

数学の世界へようこそ。この章では、基本的な数式とその表現方法について学びます。

### インライン数式

テキスト中に数式を埋め込むことができます。例えば、$E=mc^2$ はアインシュタインの有名な質量とエネルギーの等価性を表す方程式です。

### 独立した数式

より複雑な数式は、独立したブロックとして表現できます：

$$
\int_{-\infty}^{\infty} e^{-x^2} dx = \sqrt{\pi}
$$


これはガウス積分として知られる式です。

## 1.2 Juliaを使った計算例

Juliaを使って、数学的な計算を行うこともできます。以下は簡単な例です。

```julia
# 2次方程式 ax^2 + bx + c = 0 の解を求める関数
function quadratic_formula(a, b, c)
    discriminant = b^2 - 4a*c
    if discriminant < 0
        return "実数解なし"
    elseif discriminant == 0
        x = -b / (2a)
        return "重解: x = $x"
    else
        x1 = (-b + sqrt(discriminant)) / (2a)
        x2 = (-b - sqrt(discriminant)) / (2a)
        return "x1 = $x1, x2 = $x2"
    end
end

# 関数を使ってみる
println(quadratic_formula(1, 5, 6))  # x^2 + 5x + 6 = 0 の解
```
